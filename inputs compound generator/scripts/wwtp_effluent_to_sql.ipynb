```python
edar_data = pd.read_excel(path+'edar_data.xlsx')
edar_data.head()
```

```python
#Take columns codi_eu and nom_edar and convert to dictionary with key nom_edar and value codi_eu
edar_dict = edar_data.set_index('nom_edar')['codi_eu'].to_dict()
```

```python
#Functions that read the data from the excel file and return a dataframe, parses column fecha as date
def read_data(file, edar_dict):
    df = pd.read_excel(file, parse_dates=['fecha'])
    if 'cod_eu' not in df.columns:
        df['cod_eu'] = df['nom_edar'].map(edar_dict)
    #If missing value in code_eu, take column name and look for it in edar_dict
    df['cod_eu'] = df['cod_eu'].fillna(df['nom_edar'].map(edar_dict))
    #Drop id column if exists
    if 'id' in df.columns:
        df.drop('id', axis=1, inplace=True)
    return df
```

```python
prtr_edars_2 = read_data(path+'prtr_edars_2.xlsx', edar_dict)
prtr_edars_2.head()
```

```python
prtr_edars_1 = read_data(path+'prtr_edars.xlsx', edar_dict)
prtr_edars_1.head()
```

```python
#Add rows from prtr_edars_2 to prtr_edars
prtr_edars = pd.concat([prtr_edars_1, prtr_edars_2], ignore_index=True)

prtr_edars.head()
```

```python
str_columns = ['cod_eu', 'cod_aca', 'nom_edar', 'fecha', 'codi_eu']
#If column is not in str_columns, divide by 1000 (convert to micrograms/L to milligrams/L)
prtr_edars.loc[:, ~prtr_edars.columns.isin(str_columns)] = prtr_edars.loc[:, ~prtr_edars.columns.isin(str_columns)].div(1000)
prtr_edars
```

```python
#From rows that has same cod_eu and date, convert to one row with one substance per column
scarce_effluent_rows = scarce_effluent.pivot_table(index=['cod_eu', 'fecha'], columns='Substance name', values='Value').reset_index()
scarce_effluent_rows
```
